```python
import requests
import csv

headers = []
downloadedFiles = set({})

with open("data/dataset2.csv", "w") as datasetFile:
    datasetWriter = csv.writer(datasetFile)
    with open("DatasetURLs.txt", "r") as urls:
        for url in urls:
            if url in downloadedFiles:
                print(f"Duplicate file {url}")
                break
            print(f"Downloading {url}")
            r = requests.get(url)
            currentReader = iter(csv.reader(r.content.decode('utf-8').splitlines(), delimiter=','))
            header = next(currentReader)
            # Sometimes _ characters are used instead of spaces for header fields. Fix here.
            for idx, field in enumerate(header):
                header[idx] = field.replace("_", " ")
            headers.append(header)
            if len(headers) == 1:
                datasetWriter.writerow(header)
            for row in currentReader:
                datasetWriter.writerow(row)
            downloadedFiles.add(url)

for header in headers[1:]:
    if not header == headers[0]:
        print(headers[0])
        print(header)
```
